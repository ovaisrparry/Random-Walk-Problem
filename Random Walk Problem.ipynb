```python
import random
import matplotlib.pyplot as plt

num_steps = 100
num_simulations = 10

for j in range(num_simulations):

    position = 0
    positions = [0]

    for i in range(num_steps):
        if random.random() < 0.5:
            position += 1
        else:
            position -= 1
        positions.append(position)

    plt.plot(positions, label="Simulation {}".format(j+1))

plt.title("Random Walk in the x Direction")
plt.xlabel("Time Step")
plt.ylabel("Position")
plt.legend()

plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

num_steps = 100
num_simulations = 10

rms_distances = []

for j in range(num_simulations):

    position = 0
    positions = [0]

    for i in range(num_steps):
        if random.random() < 0.5:
            position += 1
        else:
            position -= 1
        positions.append(position)

    rms_distance = np.sqrt(sum([x**2 for x in positions])/num_simulations)
    rms_distances.append(rms_distance)


plt.plot(rms_distances)



plt.title("Random Walk in the x Direction")
plt.xlabel("Number of simulations run")
plt.ylabel("RMS Value")

plt.show()
```
